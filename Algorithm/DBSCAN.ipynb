üìä **Problema: Detecci√≥n de regiones densas en un mapa de sensores**

Una empresa de monitoreo ambiental ha desplegado sensores en una zona para medir temperatura y humedad. Cada sensor env√≠a sus coordenadas en metros (X, Y). Tu tarea es identificar agrupaciones naturales de sensores que se encuentren cerca entre s√≠, y detectar sensores que est√©n aislados (ruido).

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

1. Carga los datos.

```python
df = pd.read_csv('casas.csv')
df.head()
plt.scatter(df['1'], df['6'])
```

2. Aplica DBSCAN para encontrar grupos densos de sensores.

```python
clusters = DBSCAN(eps=2, min_samples=10).fit_predict(df[['1', '6']])
print(clusters)
```

3. Grafica los cl√∫steres (diferentes colores) y resalta los puntos que sean ruido (label = -1).



```python
plt.scatter(df['1'], df['6'], c=clusters, cmap='rainbow')
plt.scatter(df['1'][clusters == -1], df['6'][clusters == -1], color='yellow', marker='*')
plt.title('Coordenadas de los sensores')
plt.xlabel('Antig√ºedad de la Construcci√≥n en A√±os')
plt.ylabel('Precio de casas en pesos [1:100,000]')
plt.grid()
```
