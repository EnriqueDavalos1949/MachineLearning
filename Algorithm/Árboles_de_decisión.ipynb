📌 Enunciado: Clasificación de clientes según si comprarán o no
Tienes un conjunto de datos que contiene características de clientes y si realizaron una compra. Debes entrenar un árbol de decisión para predecir si un cliente comprará o no.

🎯 Tareas a realizar:
1. Carga el dataset.

2. Convierte la columna Compró a valores numéricos (Sí = 1, No = 0).

3. Entrena un modelo DecisionTreeClassifier con Scikit-Learn.

4. Visualiza el árbol de decisión (opcional pero recomendado).

5. Evalúa el modelo con precisión (accuracy).

6. Predice si un nuevo cliente de:

  * Edad = 35, Ingreso = 6000
  * va a comprar o no.

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

Binarización de la columna *'Compró'*

```python
df = pd.read_csv('DT.csv')
df['Compró']
df['Compró'] = df['Compró'].replace({'Sí':1, 'No':0})
df
```

```python
x = df[['Edad', 'Ingreso_mensual']]
y = df['Compró']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
#Plotear arbol de decisión
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=['Edad', 'Ingreso_mensual'], class_names=['No', 'Sí'], filled=True)
plt.show()
```

```python

accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy}')
```

```python
NewClient = pd.DataFrame({'Edad':[35], 'Ingreso_mensual':[6000]})
predict = model.predict(NewClient)
etiquetado = {1:'Sí', 0:'No'}
print(etiquetado[predict[0]])
```
